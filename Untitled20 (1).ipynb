```python
# ===========================
#  NATURAL STYLE TRANSFER
#  TensorFlow Hub version
# ===========================

import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import PIL.Image
import tensorflow_hub as hub
from google.colab import files

# ---- Function: Load image ----
def load_image(img_path, img_size=(512, 512)):
    img = PIL.Image.open(img_path)
    img = img.resize(img_size)
    img = np.array(img) / 255.0
    img = img.astype(np.float32) # Ensure the image data is float32
    img = img[tf.newaxis, :]
    return img

# ---- Upload images ----
print("/content/drive/MyDrive/Colab Notebooks/IMG-20251126-WA0047(1).jpg")
content_upload = files.upload()

print("/content/drive/MyDrive/Colab Notebooks/IMG-20251129-WA0013.jpg")
style_upload = files.upload()

content_path = list(content_upload.keys())[0]
style_path = list(style_upload.keys())[0]

content_image = load_image(content_path)
style_image = load_image(style_path)

# ---- Load NST Model from TensorFlow Hub ----
model = hub.load(
    'https://tfhub.dev/google/magenta/arbitrary-image-stylization-v1-256/2'
)

# ---- Generate stylized image ----
stylized_output = model(
    tf.constant(content_image),
    tf.constant(style_image)
)[0]

# ---- Show results ----
plt.figure(figsize=(18, 8))

plt.subplot(1, 3, 1)
plt.title("Content Image")
plt.imshow(np.squeeze(content_image))
plt.axis('off')

plt.subplot(1, 3, 2)
plt.title("Style Image")
plt.imshow(np.squeeze(style_image))
plt.axis('off')

plt.subplot(1, 3, 3)
plt.title("Stylized Result")
plt.imshow(np.squeeze(stylized_output))
plt.axis('off')

plt.show()

# ---- Save result ----
output_image = tf.image.convert_image_dtype(stylized_output, dtype=tf.uint8)[0]
output_pil = PIL.Image.fromarray(output_image.numpy())
output_pil.save("/content/drive/MyDrive/Colab Notebooks/Hasil.png")

print("Hasil disimpan sebagai: hasil_nst_hamster_batik.png")
```
